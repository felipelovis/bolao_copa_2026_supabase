```python
import os
from dotenv import load_dotenv
from supabase import create_client

load_dotenv()

url = os.environ["SUPABASE_URL"]
key = os.environ["SUPABASE_KEY"]
supabase = create_client(url, key)

print("✅ Conectado ao Supabase")
```

```python
# =====================================================
# Adicione os participantes aqui
# O nome é só pra você identificar — não precisa
# ser igual ao que a pessoa vai digitar no cadastro.
# O que valida o acesso é EMAIL + BOLAO.
# =====================================================
participantes = [
    {"nome": "Felipe",   "email": "felipevlovison@hotmail.com", "bolao": "SM3"},
    # {"nome": "Fulano", "email": "fulano@email.com",           "bolao": "SM3"},
]

# Normaliza emails para minúsculo (evita duplicatas por case)
for p in participantes:
    p["email"] = p["email"].strip().lower()

emails = [p["email"] for p in participantes]

# Verifica quais emails já existem na tabela convites
existentes_res = supabase.table("convites").select("email").in_("email", emails).execute()
emails_existentes = {r["email"] for r in existentes_res.data}

novos = [p for p in participantes if p["email"] not in emails_existentes]
ja_cadastrados = [p for p in participantes if p["email"] in emails_existentes]

if ja_cadastrados:
    print("⚠️  Já existem (pulados):")
    for p in ja_cadastrados:
        print(f"   - {p['email']} ({p['bolao']})")

if novos:
    data = supabase.table("convites").insert(novos).execute()
    print(f"\n✅ {len(novos)} convite(s) inserido(s):")
    for p in novos:
        print(f"   - {p['email']} ({p['bolao']})")
else:
    print("\nNenhum novo convite para inserir.")
```
